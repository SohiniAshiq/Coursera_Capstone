ou will learn how to convert addresses into their equivalent latitude and longitude values.

```python
import pandas as pd
```

```python
url='https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&oldid=1011037969'
```

```python
df = pd.read_html(url)[0]
```

```python
df.head()
```

```python
filt = (df['Borough'] !='Not assigned')
```

```python
df=df[filt]
```

```python
df.head()
```

```python
df.rename(columns={'Postal Code':'PostalCode'}, inplace =True)
```

```python
df.head()
```

```python
df.reset_index()
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.head(11)
```

```python
df.shape
```

```python
df_pcode = df.groupby(['PostalCode','Borough'])['Neighbourhood']
```

```python
df_pcode.apply(','.join).reset_index()
```

```python
df_postal = df_pcode.apply(','.join).reset_index()
```

```python
df_postal.head(10)
```

```python
df_postal.set_index('PostalCode', inplace = True)
```

```python
df_postal.head()
```

```python
df_latlong = pd.read_csv('Geospatial_Coordinates.csv')
```

```python
df_latlong.head()
```

```python
df_latlong.rename(columns={'Postal Code':'PostalCode'}, inplace=True)
```

```python
df_latlong.head()
```

```python
df_latlong.set_index('PostalCode', inplace = True)
```

```python
df_latlong.head()
```

```python
df_postal.join(df_latlong)
```

```python
df_toronto = df_postal.join(df_latlong)
df_toronto
```

```python
df_toronto.reset_index(inplace=True)
df_toronto.head()
```

```python
df_toronto.head(11)
```
